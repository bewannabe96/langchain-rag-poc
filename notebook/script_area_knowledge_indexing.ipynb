```python
from dotenv import load_dotenv
load_dotenv()

import os
from bson import ObjectId
from shapely.geometry import shape
from pymongo import MongoClient
```

```python
client = MongoClient(os.environ["SESSION_MONGO_CONN_STR"])
```

```python
root_area_id = "6406e6138cd140482d062b5b"

root_area_doc = client['daytrip_area']['node'].find_one({'_id': ObjectId(root_area_id)})

area_dict = {}

cursor = client["daytrip_area"]["relation"].find({"type": "CONTAIN", "nodes.0.ref": ObjectId(root_area_id)})
for relation in cursor:
    area_doc = client['daytrip_area']['node'].find_one({'_id': relation["nodes"][1]["ref"]})
    area_dict[str(area_doc["_id"])] = {
        "name": area_doc["name"],
        "geometry": shape(area_doc["geometry"]),
    }
    
area_dict
```
